```python
import matplotlib.pyplot as plt
import models 
import numpy as np

%matplotlib widget
```

```python
pzc_V = 0
concentration_M = 1e-7
bao = models.BorukhovAndelmanOrland(concentration_M, pzc_V, 15e-10)
bao_sol = bao.getDoubleLayerModelSolution(1000e-9, 10000, 1.0)

plt.close()
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(9,4), sharex=True)
ax[0].plot(bao_sol.x, bao_sol.phi, label='Huang')
ax[0].set_xlabel(r'$x$ [nm]')
ax[0].set_ylabel(r'$\phi$ [V vs. PZC]')
ax[0].set_xlim([0, 500])

ax[1].plot(bao_sol.x, bao_sol.c_cat, '-', color='tab:blue', label='Cations')
ax[1].plot(bao_sol.x, bao_sol.c_an, '-', color='tab:red', label='Anions')
# waterax = ax[1].twinx()
# waterax.plot(bao_sol.x, bao_sol.c_sol, '--', color='k', label='Solvent')
# waterax.set_ylim([0, 60])
# waterax.legend(loc='right')
ax[1].legend(loc='upper right')
ax[1].set_ylabel('c [M]')
ax[1].set_xlabel(r'$x$ [nm]')
# ax[1].set_yscale('log')

ax[2].plot(bao_sol.x, bao_sol.eps)
ax[2].set_ylim([0, 100])
ax[2].set_ylabel(r'$\varepsilon_r$')
ax[2].set_xlabel(r'$x$ [nm]')
plt.tight_layout()

plt.savefig('insulator-interface.png')
```
