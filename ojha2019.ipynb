```python
import numpy as np
import matplotlib.pyplot as plt
import edl
import plotting

%matplotlib widget
```

```python
potentials = np.linspace(-0.7, 0.3, 300)

x_list = [0, 2, 6, 25, 100]
sol_list = []

for x in x_list:
    model = edl.ProtonLPB(x * 1e-3, 27, 8, 27, 8)
    sol = model.sweep(potentials, tol=1e-4, p_h=3)
    # sol.cap = 1/(sol.cap**(-1) + 100**(-1))
    sol_list.append(sol)
```

```python
fig, ax = plotting.plot_potential_sweep(sol_list)
plt.show()
```
