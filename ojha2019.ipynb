```python
import numpy as np
import matplotlib.pyplot as plt
import models 
import potential_sweeps as sweep
import plotting

%matplotlib widget
```

```python
potentials = np.linspace(-0.4, 0.4, 100)

x_list = [0, 2, 6, 25, 100]
sol_list = []

for x in x_list:
    simple = models.Abrashkin(1e-3 + x*1e-3, 60, 8)
    sol = sweep.numerical(simple, potentials)
    
    sol_list.append(sol)

```

```python
fig, ax = plotting.plot_potential_sweep(sol_list)
plt.show()
```
